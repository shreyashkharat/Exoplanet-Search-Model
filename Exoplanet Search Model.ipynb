Missing value Imputation

```python
miss_col = ('koi_score', 'koi_period_err1', 'koi_period_err2', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact', 'koi_impact_err1', 'koi_impact_err2', 'koi_duration_err1', 'koi_duration_err2', 'koi_depth', 'koi_depth_err1', 'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2', 'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2', 'koi_model_snr', 'koi_steff', 'koi_steff_err1', 'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2', 'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'koi_kepmag')
```

```python
for col in miss_col:
    data[col] = data[col].fillna(data[col].mean())
```

Note that the target variable here has 3 categories, hence can not be represented by a bool.
We need to make a variable which takes value 0, 1 ,2 as koi_disposition is FALSE POSITIVE, CANDIDATE, CONFIRMED respectively.

```python
data['koi_disposition'] = [0]*9564
```

```python
type(data)
```

```python
data.info()
```

```python
for i in range(0, 9564):
    if(data.at[i, 'koi_disposition_CANDIDATE'] == 1):
        data.at[i, 'koi_disposition'] = 1
    elif(data.at[i, 'koi_disposition_CONFIRMED'] == 1):
        data.at[i, 'koi_disposition'] = 2
    elif(data.at[i, 'koi_disposition_CANDIDATE'] == 0 and data.loc[i, 'koi_disposition_CONFIRMED'] == 0):
        data.loc[i, 'koi_disposition'] = 0
```

```python
data
```

```python
del data['koi_disposition_CANDIDATE']
del data['koi_disposition_CONFIRMED']
```

### Train Test split and variable formation

```python
train_data = data.loc[:6999, :]
test_data = data.loc[7000:, :]
```

```python
x_train = train_data.loc[:, train_data.columns != 'koi_disposition']
y_train = train_data['koi_disposition']
x_test = test_data.loc[:, test_data.columns != 'koi_disposition']
y_test = test_data['koi_disposition']
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape, data.shape
```

# Model Building

## Logistic Model

```python
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression(max_iter = 100000)
model_log.fit(x_train, y_train)
```

Even after a 100,000 iterations the training of model isn't as the Logistic function here is in 3 dimensions. Anyway let's evaluate it.

```python
log_train_predict = model_log.predict(x_train)
log_test_predict = model_log.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
con_log_train = confusion_matrix(y_train, log_train_predict)
con_log_test = confusion_matrix(y_test, log_test_predict)
```

```python
con_log_train
```

```python
con_log_test
```

```python
log_train_acc = accuracy_score(y_train, log_train_predict)
log_test_acc = accuracy_score(y_test, log_test_predict)
log_train_acc, log_test_acc
```

The accuracy of Logistic Model is fine, better performance can be given by other models.

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
k_hyper = {'n_neighbors': range(1, 101)}
```

```python
grid_search = GridSearchCV(KNeighborsClassifier(), k_hyper)
```

```python
from sklearn import preprocessing
scaler_train = preprocessing.StandardScaler().fit(x_train)
x_train_scaled = scaler_train.transform(x_train)
scaler_test = preprocessing.StandardScaler().fit(x_test)
x_test_scaled = scaler_test.transform(x_test)
```

```python
grid_search.fit(x_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
model_knn = grid_search.best_estimator_
```

```python
knn_train_predict = model_knn.predict(x_train_scaled)
knn_test_predict = model_knn.predict(x_test_scaled)
```

```python
con_knn_train = confusion_matrix(y_train, knn_train_predict)
con_knn_test = confusion_matrix(y_test, knn_test_predict)
```

```python
con_knn_train
```

```python
con_knn_test
```

```python
knn_train_acc = accuracy_score(y_train, knn_train_predict)
knn_test_acc = accuracy_score(y_test, knn_test_predict)
knn_train_acc, knn_test_acc
```

This model behaves well, the previous model has test set accuracy greater than train set accuracy, the main reason for this is irregular distribution of data, as these is astronomical data recorded we highly accurate instruments. Some observations might seem as outliers but there's a high probability that the observation is not a outlier but a genuine observation. Due to this the data might have a irregular distribution. The distribution doesn't affect K-Nearest Model as this model doesn't use any function but uses geometric arguements to estimate the target variable.
